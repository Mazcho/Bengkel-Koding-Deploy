```python
from imblearn.over_sampling import SMOTE
```

```python
# oversampling
smote = SMOTE(random_state=42)
X_smote_resampled, y_smote_resampled = smote.fit_resample(X, y)
```

```python

plt.figure(figsize=(12, 4))

df_nonresampled = pd.DataFrame(data=y)

plt.subplot(1, 2, 1)
countplot_nonresampled = df_nonresampled.value_counts().plot(kind='bar', figsize=(10, 6), color=sns.color_palette("plasma"))
plt.title("target before over sampling with SMOTE ")
plt.xticks(rotation=0);

# Tambahkan bar label untuk setiap bar di plot pertama
for container in countplot_nonresampled.containers:
    plt.bar_label(container, fmt='%d')

plt.subplot(1, 2, 2)
df_resampled = pd.DataFrame(data=y_smote_resampled)
countplot_resampled = df_resampled.value_counts().plot(kind='bar', figsize=(10, 6), color=sns.color_palette("plasma"))
plt.title("target after over sampling with SMOTE")
plt.xticks(rotation=0);

# Tambahkan bar label untuk setiap bar di plot kedua
for container in countplot_resampled.containers:
    plt.bar_label(container, fmt='%d')

plt.tight_layout()
plt.show()
```

### Oversample + Normalisasi + Tunning

#### Logistic Regression

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

# Create Logistic Regression model
logistic_model = LogisticRegression()

# Define parameter grid for tuning
param_grid = {
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 200, 300, 400],
    'C': uniform(loc=0, scale=4)
}

# Create RandomizedSearchCV object with lower n_iter
logistic_model_tuned = RandomizedSearchCV(logistic_model, param_distributions=param_grid, n_iter=5, cv=5, n_jobs=-1)

# Fit the model to the data
logistic_model_tuned.fit(X_train_normal, y_train_normal)

# Get the best parameters
best_params = logistic_model_tuned.best_params_
print(f"Best parameters: {best_params}")

# Make predictions on the test set
y_pred_logistic = logistic_model_tuned.predict(X_test_normal)

# Evaluate the accuracy
accuracy = accuracy_score(y_test_normal, y_pred_logistic)
print(f"Accuracy on the test set: {accuracy:.2f}")
```

```python
# Evaluate the LogisticRegression model
print("Logistic Regression Model:")
accuracy_logreg_smote_normal_tuned = round(accuracy_score(y_test_normal,y_pred_logistic),3)
print("Accuracy:", accuracy_logreg_smote_normal_tuned)
print("Classification Report:")
print(classification_report(y_test_normal, y_pred_logistic))
```

```python
cm = confusion_matrix(y_test_normal, y_pred_logistic)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title('Confusion Matrix')
plt.xlabel('True')
plt.ylabel('Predict')
plt.show()
```

#### AdaBoost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Create AdaBoostClassifier model
adaboost_model = AdaBoostClassifier()

# Define parameter grid for tuning
param_grid = {
    'n_estimators': randint(5, 50),  # Reducing the number of trees further
    'learning_rate': [0.5],  # Adjusting learning rate
    'algorithm': ['SAMME', 'SAMME.R'],
}

# Create RandomizedSearchCV object
adaboost_model_tuned = RandomizedSearchCV(adaboost_model, param_distributions=param_grid, n_iter=5, cv=5, n_jobs=-1)

# Fit the model to the data
adaboost_model_tuned.fit(X_train_normal, y_train_normal)

# Get the best parameters
best_params_adaboost = adaboost_model_tuned.best_params_
print(f"Best parameters for AdaBoostClassifier: {best_params_adaboost}")

# Make predictions on the test set
y_pred_adaboost = adaboost_model_tuned.predict(X_test_normal)

# Evaluate the accuracy
accuracy_adaboost = accuracy_score(y_test_normal, y_pred_adaboost)
print(f"Accuracy on the test set for AdaBoostClassifier: {accuracy_adaboost:.2f}")
```

```python
# Evaluate the LogisticRegression model
print("Logistic Regression Model:")
accuracy_ada_smote_normal_tuned = round(accuracy_score(y_test_normal,y_pred_adaboost),3)
print("Accuracy:", accuracy_ada_smote_normal_tuned)
print("Classification Report:")
print(classification_report(y_test_normal, y_pred_adaboost))
```

```python
cm = confusion_matrix(y_test_normal, y_pred_adaboost)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title('Confusion Matrix')
plt.xlabel('True')
plt.ylabel('Predict')
plt.show()
```

#### Multilayer Perceptron (MLP)

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import RandomizedSearchCV

# Create MLP Classifier model
mlp_model = MLPClassifier()

# Define parameter grid for tuning
param_grid = {
    'activation': ['logistic', 'relu'],
    'solver': ['sgd', 'adam'],
    'learning_rate': ['constant', 'invscaling', 'adaptive'],
}

# Create RandomizedSearchCV object
mlp_model_tuned = RandomizedSearchCV(mlp_model, param_distributions=param_grid, n_iter=10, cv=5, n_jobs=-1)

# Fit the model to the data
mlp_model_tuned.fit(X_train_normal, y_train_normal)

# Get the best parameters
best_params_mlp = mlp_model_tuned.best_params_
print(f"Best parameters for MLP Classifier: {best_params_mlp}")

# Make predictions on the test set
y_pred_mlp = mlp_model_tuned.predict(X_test_normal)

# Evaluate the accuracy
accuracy_mlp = accuracy_score(y_test_normal, y_pred_mlp)
print(f"Accuracy on the test set for MLP Classifier: {accuracy_mlp:.2f}")
```

```python
# Evaluate the LogisticRegression model
print("Logistic Regression Model:")
accuracy_mlp_smote_normal_tuned = round(accuracy_score(y_test_normal,y_pred_mlp),3)
print("Accuracy:", accuracy_mlp_smote_normal_tuned)
print("Classification Report:")
print(classification_report(y_test_normal, y_pred_mlp))
```
