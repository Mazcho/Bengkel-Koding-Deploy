```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [4, 2, 6, 9, 10, 22]
y = [0, 1, 2, 3, 4, 5]
z = [7, 2, 5, 3, 8, 1]
plt.title("testing")
plt.xlabel("angka random")
plt.ylabel(["angka urut", "angka random"])
plt.plot(x,y,z)
```

```python
fruit = ['apple', 'blueberry', 'anggur', 'jeruk']
count = [50, 100, 30, 25]
bar_colors = ['tab:red', 'tab:blue', 'tab:purple', 'tab:orange']

fig,ax = plt.subplots()
ax.bar(fruit, count, color=bar_colors)
ax.set_title('jumlah setiap buah')
ax.set_ylabel('jumlah')
plt.show()
```

```python
label = ['Frogs', "tigers", "dogs", "cats"]
size = [15.2, 29.8, 30.5, 25.5]

fig1, ax1 = plt.subplots()
ax1.pie(size, labels=label, autopct='%.1f%%')
plt.title ("pie chart hewan")
plt.legend(label)

plt.show()
```
