# Data Collection and Data understanding

Data set pada projek ini diambil dari https://archive.ics.uci.edu/dataset/45/heart+disease.

Direktori ini berisi 4 database mengenai diagnosis penyakit jantung.
   Semua atribut bernilai numerik. Data dikumpulkan dari
   empat lokasi berikut:

     1. Yayasan Klinik Cleveland (cleveland.data)
     2. Institut Kardiologi Hongaria, Budapest (hungarian.data)
     3.V.A. Pusat Medis, Long Beach, CA (long-beach-va.data)
     4. Rumah Sakit Universitas, Zurich, Swiss (switzerland.data)

   Setiap database memiliki format instance yang sama. Sedangkan databasenya ada 76
   atribut mentah, hanya 14 yang benar-benar digunakan. Atribut yang digunakan diantaranya
   
1. #3 (age): usia
2. #4 (sex): Jenis Kelamin
3. #9 (cp): jenis penyakit dada

5. #12 (chol): kolesterol serum dalam mg/dl
6. #16 (fbs): (gula darah puasa >120 mg/dl)
7. #19 (restecg): hasil elektrokardiografi istirahat
8. #32 (thalach) : detak jantung maksimal tercapai
9. #38 (exang): angina akibat olahraga
10. #40 (oldpeak): Depresi ST yang disebabkan oleh olahraga versus istirahat
11. #41 (kemiringan): kemiringan latihan segmen ST puncak
12. #44 (ca): jumlah bejana besar (0-3) yang diwarnai dengan flourosopy
13. #51 (thal): 3 = biasa; 6 = cacat tetap; 7 = cacat reversibel
14. #58 (num): diagnosis penyakit jantung (status penyakit angiografi)

# Load Dataset

Untuk memuat data, kita lakukan import library terlebih dahulu. Import ini ditujukan untuk memasukan pustaka dalam menanangani pemrosesan dalam pengolahan data. 

```python
#Import library 
import pandas as pd
import numpy as np
import re
import itertools
```

```python
#load dataset
dir = "hungarian.data"
```

```python
#melakukan encoding pada dataset
with open(dir, encoding='Latin1') as file:
    lines = [line.strip() for line in file]
    
#membuatnya kedalam data frame
data = itertools.takewhile(
    lambda x: len(x) == 76,
    (' '.join(lines[i:(i + 10)]).split() for i in range(0, len(lines), 10)))

df = pd.DataFrame.from_records(data)
df.head()
```

```python
# Melihat informasti tipe data yang ada serta jumlah data yang ada setiap vaiabel
df.info()
```

Sesuai dengan arahan pada sumber dataset, terdapat kolom nama 75 berisikan nilai name. Name mewakili ahkir dari setiap baris data, dan nilai -9.0 mewakili missing value. Maka dari itu kita perlu menghapus kolom 75 dan mengubah tipe data object menjadi float agar kita dapat mengubah -9.0 menjadi Nan

```python
#Menghapus kolom 75 berisikan name
df.drop(75,axis=1,inplace=True)

#menhapus kolom 0 karena itu index data
df.drop(0,axis=1,inplace=True)
```

```python
#mengubah tipe data kolom object menjadi float
df = df.astype(float)
```

```python
#melihat informasi tipe data 
df.info()
```
